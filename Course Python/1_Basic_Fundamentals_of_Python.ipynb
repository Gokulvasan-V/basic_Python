## **7. Type Casting (or) Type Conversion**

Type conversion is the process of converting a data type into another data type.

**Two Types of Type Casting**

1. **Implicit** --> Python automatically converts one data type to another data type without any user's need; quite unique feature of Python language.

2. **Explicit** --> via user intervention or manually as per the requirement.



image.png

**NOTE:**

* Inside the string if we have a values like numbers, we can convert it as an integer or float (Which we want).

* At the same time, Inside the string if we have a values like character, we can't convert it as an integer or float.

````python
# Implicit type conversion

a = 23
b = float(a)

print(a, type(a))
print(b, type(b))
````

````python
a = input()
print(a)
print(type(a)) # 'a' is implicitly converted to string
````

````python
x = 5
y = 2.5
z = x + y  # Integer `x` is implicitly converted to float
print(z)
````

````python
c = 2,3,4,5
print(c, type(c))
````

````python
x = True
y = 5
z = x + y  # Boolean `x` is implicitly converted to integer (True -> 1)
print(z)
````

````python
# Explicit type conversion

a = int(input("'a' value is"))
print(a, type(a))

b = float(input("'b' value is"))
print(b, type(b))
````
