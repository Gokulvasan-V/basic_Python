# **conditional statements**

```python
a= 200
b=500
if a<b:
  print("b is greater then a") # here we can use if only
```

```python
if a>b:
  print("a is less then b")
else:
  print("b is greater then a") # here we use if and else
```

```python
c=500
if a>b:
  print("a is less then b")
elif b==c:
  print('b equal to c')
else:
  print("b is greater then a") # here we use if else as well as elif
```
