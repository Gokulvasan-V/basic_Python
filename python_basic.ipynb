## **List Comprehension**

```python
list_1=['apple','banana','mango','watermelon','graps','Plum', 'Mulberry']
list_new=[]
```

```python
#tradition method
for i in list_1:
  if 'a' in i:
    list_new.append(i)

list_new
```

```python
list_new_2=[i for i in list_1 if 'a' in i]
list_new_2
```
